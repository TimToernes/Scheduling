## Data import 

```python
df_pv = pd.read_csv('data/pv_optimal.csv',sep=';')
df_pv.index = pd.DatetimeIndex(df_pv['utc_time'])
df_onshore = pd.read_csv('data/onshore_wind_1979-2017.csv',sep=';')
df_onshore.index = pd.DatetimeIndex(df_onshore['utc_time'])
df_el = pd.read_csv('data/electricity_demand.csv',sep=';')
df_el.index = pd.DatetimeIndex(df_el['utc_time'])
```

```python
def annuity(n,r):
    """Calculate the annuity factor for an asset with lifetime n years and
    discount rate of r, e.g. annuity(20,0.05)*20 = 1.6"""
    if r > 0:
        return r/(1. - 1./(1.+r)**n)
    else:
        return 1/n
```

```python
capital_cost_OCGT=annuity(25,0.07)*560000*(1+0.033) # in €/MW
fuel_cost=200   #21.6 # in €/MWh_th
efficiency=0.39
marginal_cost_OCGT=fuel_cost/efficiency # in €/MWh_el

capital_cost_onshorewind=annuity(30,0.07)*910000*(1+0.033) # in €/MW

capital_cost_solar=annuity(25,0.07)*425000*(1+0.03) # in €/MW
```

## Pypsa model

```python
t_start = '2015-01-01T00:00Z'
t_end = '2015-03-01T23:00Z'

hours=pd.date_range(t_start,t_end,freq='H') #'2017-12-31T23:00Z'
network = pypsa.Network()
network.set_snapshots(hours)
```

```python
#add three buses
buses = ['DNK','NOR','SWE']
n_buses = len(buses)
for i in range(len(buses)):
    network.add("Bus",buses[i],
                #v_nom=20.,
               x = i,
               y = i%2)
```

```python
#add three lines in a ring
for i in range(len(buses)):
    network.add("Link","My line {}".format(i),
                bus0=buses[i],
                bus1=buses[(i+1)%n_buses],
                #x=0.1,
                #r=0.3,
                length = 100,
                p_nom_extendable=True,
                p_min_pu=-1,
                capital_cost=75753)
```

```python
network.plot()
```

```python
network.add("Carrier","gas",
      co2_emissions=0.2)

# Add dispatchable generator
for i in range(n_buses):
    network.add("Generator","{} ocgt".format(buses[i]),
                type = 'ocgt',
                bus=buses[i],
                committable=True,
                capital_cost=capital_cost_OCGT,
                marginal_cost=marginal_cost_OCGT,
                p_nom_extendable=True,
                carrier="gas",)
```

```python
# Add wind 
for i in range(n_buses):
    network.add("Generator","{} wind".format(buses[i]),
                type = 'wind',
                bus=buses[i],
                marginal_cost=0,
                capital_cost = capital_cost_onshorewind,
               committable=True,
               p_nom_extendable=True,
               p_max_pu=df_onshore.loc[t_start:t_end,buses[i]]) 
```

```python
# Add solar
for i in range(n_buses):
    network.add("Generator","{} solar".format(buses[i]),
                bus=buses[i],
                type = 'solar',
               marginal_cost=0,
                capital_cost = capital_cost_solar,
               committable=True,
               p_nom_extendable=True,
               p_max_pu=df_pv.loc[t_start:t_end,buses[i]]) 
```

```python
# Add loads
for i in range(n_buses):
    network.add("Load","load {}".format(i),
                bus=buses[i],
                p_set=df_el.loc[t_start:t_end,buses[i]])
```

## Sceduling
The goal of the sceduling algorithm is to minimize a) price paid for the electricity consumed, and b) CO2 intensity in the electricity consumed. The objective function can then be formulated as  
$\text{minimize} \; f(x) = \sum_{t} x_t(\alpha*p_t + (1-\alpha)c_t) $   
where $x_t$ is the energy consumed in the given hour $t$, $p_t$ is the forecasted electricity price, $c_t$ is the forecasted CO2 intesity and $\alpha$ is the priorization ratio between low electricity price and low CO2 intesity. 

The technology to be scheduled for may be subject to a ramp constraint, limiting the hourly change in production. The constraint can be formulated as follows:  
$|x_{t+1}-x_t|< r*P $  
where $r$ is the maximum allowable rate given in %, and P is the installed capacity of that specific technology. 

A constraint on the amount of full load hours $\text{flh}$ deliverd in a given time frame $T$ is also introduced. $\sum_{t\in T} \frac{x_t}{P} = \text{flh} $ 

```python
spot_price = [140]
max_rate = 50
rate = 0
for i in range(len(hours)-1):
    rate = rate + (np.random.rand()*2-1)*max_rate
    spot_price.append(spot_price[i-1]+rate)
    
power_price = pd.Series(index=hours,data=spot_price)
```

```python
CO2_int = [100]
max_rate = 50
rate = 0 
for i in range(len(hours)-1):
    rate = rate + (np.random.rand()*2-1)*max_rate
    CO2_int.append(CO2_int[i-1]+rate)


CO2_intesity = pd.Series(index=hours,data=CO2_int)
```

```python
t_start = '2015-01-01T00:00Z'
t_end = '2015-01-01T23:00Z'

hours=pd.date_range(t_start,t_end,freq='H') #'2017-12-31T23:00Z'
network = pypsa.Network()
network.set_snapshots(hours)
```

```python
alpha = 0.5

marginal_cost = power_price*alpha + CO2_intesity*(1-alpha)
```

```python
#marginal_cost = df_el.loc[t_start:t_end,buses[i]]
```

```python
network.add('Bus','H2')
```

```python
network.add('Generator','Electrolyzer',
            bus='H2',
            committable=True,
            p_nom=1,
            p_nom_extendable=False,
            marginal_cost=marginal_cost,
            p_min_pu = 0.1,
            min_up_time = 2,
            ramp_limit_up = 0.2,
            ramp_limit_down = 0.2,
            ramp_limit_start_up = 0.1,
           )
```

```python
flh = 10

load = pd.Series(index=hours,data=np.zeros(len(hours)))
load.iloc[-1] = flh 
```

```python
network.add('Load','demand',
            bus='H2',
            p_set=load)
```

```python
network.add('Store','store',
           bus='H2',
           e_nom=flh)
```

```python
network.lopf(solver_name="gurobi")
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=hours,y=network.generators_t.p['Electrolyzer'],name='production'))
fig.add_trace(go.Scatter(x=hours,y=marginal_cost/max(marginal_cost),name='cost'))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=hours,y=CO2_intesity,name='CO2 intensity'))
fig.add_trace(go.Scatter(x=hours,y=power_price,name='power price'))
fig.add_trace(go.Scatter(x=hours,y=marginal_cost,name='marginal'))
```
